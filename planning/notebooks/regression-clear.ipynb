```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os

import plots
```

### 3 blocks

```python
df = pd.read_csv('../results/blocksworld/p-clear-3blocks-0-flag/regression.csv')
df.total_features.unique()
```

```python
df = df[df.method == 'L0']
df["complexity"] = 0
features = 4
df_short = df[df.total_features==features].sort_values(["parameter", "score", "complexity"], ascending=(True, False, True))
plots.max_score_by_parameter(df_short, features)
plots.max_score_by_num_features(df, step=100)
plots.support_by_lambda(df, features)
```

```python
df.iloc[df.score.argmax()].features_names, df.iloc[df.score.argmax()].feature_weights
```
