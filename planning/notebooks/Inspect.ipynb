## $On(x,y)$

$V(s) = 2 x(s) + 2 y(s) - 2g(s) - 2X(s) + H(s) + 2$

$x(s) \rightarrow $ number of blocks above x

$y(s) \rightarrow $ number of blocks above y

$g(s) \rightarrow $ block x is on y

$X(s) \rightarrow $ block x is being held

$H(s) \rightarrow $ holding any other block than X
