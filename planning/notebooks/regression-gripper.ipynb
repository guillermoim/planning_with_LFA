```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os

import plots
```

## Instance 1

```python
df = pd.read_csv('../results/gripper/prob01_1-flag/regression.csv')
print(df.total_features.unique().tolist())

df = df[df.method == 'L0']
df["complexity"] = 0
features = 5
df_short = df[df.total_features==features].sort_values(["parameter", "score", "complexity"], ascending=(True, False, True))
plots.max_score_by_parameter(df_short, features)
plots.max_score_by_num_features(df, step=100)
plots.support_by_lambda(df, features)
```
